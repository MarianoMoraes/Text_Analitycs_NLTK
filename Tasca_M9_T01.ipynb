#### Analizaremos una review sobre productos de Amazon, en este caso un Apple Watch Series 9.

```python
review = "I recently purchased the Apple Watch on Prime Day and it has exceeded all my expectations! It's an absolute game-changer, earning a solid five stars from me.  The sleek design is a standout feature—elegant, modern, and incredibly comfortable to wear all day. The display is vibrant and sharp, making it easy to read notifications, messages, and health data at a glance. The array of features is mind-blowing. From tracking my workouts and heart rate to receiving calls and messages, this watch seamlessly integrates with my lifestyle. The fitness tracking capabilities are especially impressive, motivating me to stay active and reach my goals. Battery life is fantastic, lasting through a whole day and more, and the charging is quick and hassle-free. The ease of use and intuitive interface make navigating through apps and functions a breeze. All in all, the Apple Watch is an exceptional piece of technology that has truly enhanced my daily routine. If you're looking for a smartwatch that's stylish, functional, and reliable, this is the one to go for!"
print(review)
```

#### Primero pasamos todas las letras a minusculas

```python
review = review.lower()
print(review)
```

#### A continuacion tokenizaremos las palabras una a una.

```python
from nltk.tokenize import word_tokenize

tokenized_words = nltk.word_tokenize(review)
tokenized_words
```

#### Seguidamente contaremos las palabras para obtener su frecuencia.

```python
from nltk import FreqDist
words_frecuency = FreqDist(tokenized_words)
print(words_frecuency.most_common(10))
```

#### Vemos que se estan contando las comas, puntos y espacios por lo que procederemos a eliminarlas.

```python
tokens_cleaned = [token for token in tokenized_words if token.isalpha()]
tokens_cleaned
```

```python
from nltk import FreqDist

words_frecuency = FreqDist(tokens_cleaned)

print(words_frecuency.most_common(10))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Crea una cadena de texto a partir de la lista de palabras tokenizadas sin stopwords
texto_para_wordcloud = ' '.join(tokens_cleaned)

# Configura el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto_para_wordcloud)

# Muestra el WordCloud utilizando Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# - Exercici 2
### Treu les stopwords i realitza stemming al teu conjunt de dades.

#### A continuación utilizaremos el metodo stop_words para eliminar las palabras paso como pueden ser (y, es, el, lo, la, etc)

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))  # Puedes cambiar 'spanish' por otro idioma si es necesario

tokens_stop_words = [token for token in tokens_cleaned if token.lower() not in stop_words]

tokens_stop_words
```

#### Aplicamos nuevamente el contador de frecuencia

```python
from nltk import FreqDist

words_frecuency = FreqDist(tokens_stop_words)

print(words_frecuency.most_common(10))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Crea una cadena de texto a partir de la lista de palabras tokenizadas sin stopwords
texto_para_wordcloud = ' '.join(tokens_stop_words)

# Configura el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto_para_wordcloud)

# Muestra el WordCloud utilizando Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### Continuamos aplicando el metodo de steamming para reducir las palabras a su formato base o de raiz

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in tokens_stop_words]
stemmed_tokens
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Crea una cadena de texto a partir de la lista de palabras tokenizadas sin stopwords
texto_para_wordcloud = ' '.join(stemmed_tokens)

# Configura el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(texto_para_wordcloud)

# Muestra el WordCloud utilizando Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sentimiento_review = ' '.join(stemmed_tokens)

# Crear una instancia del analizador de intensidad de sentimiento (VADER)
sia = SentimentIntensityAnalyzer()

# Obtener la puntuación de polaridad para el texto
score_sentimiento = sia.polarity_scores(sentimiento_review)

print(score_sentimiento)
```

#### Observando el score obtenido luego de aplicar el metodo VADER, vemos que la propocion de nuestra review tiene 0 puntos como negativo, 0.854 neutral, y 0.146 como positivo. Por ultimo Compuound con un 0.90 es es la puntuacion de polaridad compuesta por las anteriores 3 variables. En nuestro caso el modelo indica que nuestro texto tiene una polaridad positiva.
